```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
```

```python
# initiate state

class CricketState(TypedDict):
    runs: int
    balls: int
    fours: int
    sixes: int

    sr: float
    bpb: float
    boundary_percent: float
    summary: str
```

```python
# define calculat strike rate function
def calculate_sr(state: CricketState):
    sr = (state['runs']/state['balls'])*100
    
    return {"sr": sr}
```

```python
# define calculate balls per boundary function

def calculate_bpb(state: CricketState):
    bpb = state["balls"]/(state["fours"] + state["sixes"])

    return {"bpb", bpb}
```

```python
# define calculate boundary percentage function

def calculate_boundary_percent(state: CricketState):
    boundary_percent = (state['runs']/(state['fours']*4 + state['sixes']*6))*100

    return {"boundary_percent", round(boundary_percent, 2)}



    
```

```python
def summary(state: CricketState):
    summary = f"Strike Rate: {state['sr']} \n Balls per boundary: {state['bpb']} \n Boundary percentage: {state['boundary_percent']}"
    return{"summary": summary}
```

```python
# graph, nodes, edges

graph = StateGraph(CricketState)

graph.add_node("calculate_sr", calculate_sr)
graph.add_node("calculate_bpb", calculate_bpb)
graph.add_node("calculate_boundary_percent", calculate_boundary_percent)
graph.add_node("summary", summary)

graph.add_edge(START, calculate_sr)
graph.add_edge(START, calculate_bpb)
graph.add_edge(START, calculate_boundary_percent)

graph.add_edge(calculate_sr, summary)
graph.add_edge(calculate_bpb, summary)
graph.add_edge(calculate_boundary_percent, summary)

graph.add_edge(summary, END)

graph.compile()


```
