```python
from langgraph.graph import StateGraph, START, END
from langchain_ollama import ChatOllama
from pydantic import Field, BaseModel
from typing import TypedDict, Annotated
import operator
```

```python
# model for chatting
model = ChatOllama(model="llama3.2")
```

```python
class StructuredOutput(BaseModel):
    feedback: str = Field(description= "Detailed feedback for the essay given")
    score: int = Field(description= "Rate the essay out of 10", le = 10, ge = 0)
```

```python
structured_model = model.with_structured_output(StructuredOutput)
```

```python
# Defining state 

class UPSCState(TypedDict):
    essay: str
    language_feedback: str
    analysis_feedback: str
    clarity_feedback: str
    overall_feedback: str

    scores: Annotated[list[int], operator.add]
    avg_score: float
```

```python
## define evaluate language function for the graph

def evaluate_language(state: UPSCState):
    prompt = f"Evaluate the language quality of this essay and provide a feedback and give score for this essay out of 10 \n {state['essay']}"
    output = structured_model.invoke(prompt)
    return {"language_feedback": output.feedback, "scores": [output.score]}
```

```python
## define evaluate analysis function for the graph

def evaluate_analysis(state: UPSCState):
    prompt = f"Evaluate the depth of the analysis of this essay and provide a feedback and give score for this essay out of 10 \n {state['essay']}"
    output = structured_model.invoke(prompt)
    return {"analysis_feedback": output.feedback, "scores": [output.score]}
```

```python
## define evaluate analysis function for the graph

def evaluate_clarity(state: UPSCState):
    prompt = f"Evaluate the clarity of of this essay and provide a feedback and give score for this essay out of 10 \n {state['essay']}"
    output = structured_model.invoke(prompt)
    return {"clarity_feedback": output.feedback, "scores": [output.score]}
```

```python
def overall_feedback(state: UPSCState):
    prompt = f"Using the feedbacks of evaluation of response through language, carity of thought, depth of analysis content give me overall content \n Language feedback: {state['language_feedback']} \n Depth of analysis feedback: {state['analysis_feedback']} \n Clarity of thought feedback: {state['clarity_feedback']}"

    output = model.invoke(prompt)
    avg_score = sum(state['scores'])/ len(state["scores"])
    return {"overall_feedback": output.content, "avg_score": avg_score}
```

```python
# graph

graph = StateGraph(UPSCState)

# nodes
graph.add_node("evaluate_language", evaluate_language)
graph.add_node("evaluate_analysis", evaluate_analysis)
graph.add_node("evaluate_clarity", evaluate_clarity)
graph.add_node("overall_feedback", overall_feedback)


# edges
graph.add_edge(START, "evaluate_language")
graph.add_edge(START, "evaluate_analysis")
graph.add_edge(START, "evaluate_clarity")

graph.add_edge("evaluate_language", "overall_feedback")
graph.add_edge("evaluate_analysis", "overall_feedback")
graph.add_edge("evaluate_clarity", "overall_feedback")

graph.add_edge("overall_feedback", END)

# compile the graph
workflow = graph.compile()
```

```python
input_state = UPSCState(essay = essay)
final_state = workflow.invoke(input_state)
```
