```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict, Annotated
```

```python
# state 
class QuadraticState(TypedDict):
    a: int
    b: int
    c: int

    equation: str
    d: float
    result : str
```

```python
def show_equation(state: QuadraticState):
    #ax**2 + b*x + c
    a = state["a"]
    b = state["b"]
    c = state["c"]

    equation = f"{a}x**2 + ({b})*x + ({c})"
    return {"equation": equation}

# finding descriminant
def find_descriminant(state: QuadraticState):
    a = state["a"]
    b = state["b"]
    c = state["c"]

    d = b**2 - 4*a*c

    return {"d": d}

def real_roots(state: QuadraticState):
    a = state["a"]
    b = state["b"]
    d = state["d"]

    root1 = (-b + d**0.5)/ (2*a)
    root2 = (-b - d**0.5)/ (2*a)

    result = f"Two roots for the equation are {root1} and {root2}"


    return {"result": result}

def equal_roots(state: QuadraticState):
    a = state["a"]
    b = state["b"]
    root = (-b)/ (2*a)
    result = f"One real root is {root}"

    return {"result": result}


def no_real_roots(state: QuadraticState):
    result = f"There are no real roots for this equation"
    return {"result": result}

def conditional_check(state: QuadraticState):
    d = state["d"]
    if d > 0:
        return "real_roots"
    elif d == 0:
        return "equal_roots"
    else: 
        return "no_real_roots"
    


```

```python
# graph
graph = StateGraph(QuadraticState)

#nodes
graph.add_node("show_equation", show_equation)
graph.add_node("find_descriminant", find_descriminant)
graph.add_node("real_roots", real_roots)
graph.add_node("equal_roots", equal_roots)
graph.add_node("no_real_roots", no_real_roots)

graph.add_edge(START, "show_equation")
graph.add_edge("show_equation", "find_descriminant")
graph.add_conditional_edges("find_descriminant", conditional_check)
graph.add_edge("real_roots", END)
graph.add_edge("equal_roots", END)
graph.add_edge("no_real_roots", END)

workflow = graph.compile()
```

```python
initial_state = QuadraticState(a = 20, b = 2, c = 1)
final_state = workflow.invoke(initial_state)
```
