```python
# Splits just by characters. You can specify the separator, chunk size and chunk overlap.
from langchain_text_splitters import CharacterTextSplitter
```

```python
text = """On a square of dust and emerald green,
Where willow meets a leathered dream,
The pitch lies patient, scored and worn,
Awaiting duels each gentle morn.

The bowler runs with thundered breath,
A measured dance with speed and death,
The ball swings late, a whispered plea,
To bend the laws of gravity."""
```

```python
splitter = CharacterTextSplitter(separator = "", chunk_size = 100, chunk_overlap = 0)
```

```python
chunks = splitter.split_text(text)
```
