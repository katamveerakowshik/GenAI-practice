```python
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint
from dotenv import load_dotenv
load_dotenv()
```

```python
def ask_llm(prompt): 
    llm = HuggingFaceEndpoint(
    repo_id="meta-llama/Llama-3.2-1B-Instruct",
    task="text-generation")

    chat = ChatHuggingFace(llm=llm)
    result = chat.invoke(prompt)
    return result.content
```

```python
#Zero-Shot Prompting
ask_llm("Summarise the plot of the movie Inception in one sentence.")
```

```python
#Few-Shot Prompting, giving examples in the prompt
prompt = """
Translate English to Pirate Speak:
English: Hello friend
Pirate: Ahoy matey

English: Where is the treasure?
Pirate:"""
print(ask_llm(prompt))
```

```python
#Role Based prompting, assigning a role to the LLM
prompt = """
You are a helpful math tutor. 
Explain how to solve 16 รท 4 to a 5-year-old.
"""
print(ask_llm(prompt))
```

```python
#Chain of Thought Prompting, encouraging step by step reasoning
prompt = """
Solve: A train leaves at 3pm traveling 60 mph. Another leaves at 4pm at 90 mph. 
When will the faster train catch up? 
Think step by step.
"""
print(ask_llm(prompt))
```
