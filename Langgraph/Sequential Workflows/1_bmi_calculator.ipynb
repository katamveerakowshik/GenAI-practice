```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
```

```python
# state for the graph
class BMIState(TypedDict):
    weight: float
    height: float
    bmi: float
    label: str
```

```python
# add bmi_calculator node (function)

def bmi_calculator(state: BMIState) -> BMIState:
    weight = state["weight"]
    height = state["height"]

    bmi = weight / (height ** 2)
    state["bmi"] = bmi
    return state
```

```python
# add label bmi node (function)

def label_bmi(state: BMIState) -> BMIState:
    bmi = state["bmi"]

    if bmi < 18.5:
        label = "Underweight"
    elif 18.5 <= bmi < 25:
        label = "Normal weight"
    elif 25 <= bmi < 30:
        label = "Overweight"
    else:
        label = "Obese"

    state["label"] = label
    return state
```

```python
# define the grpah, nodes and edges
graph = StateGraph(BMIState)

# add nodes
graph.add_node("bmi_calculator", bmi_calculator)
graph.add_node("label_bmi", label_bmi)

# add edges
graph.add_edge(START, "bmi_calculator")
graph.add_edge("bmi_calculator", "label_bmi")
graph.add_edge("label_bmi", END)

# execute the graph
workflow = graph.compile()
```

```python
# invoke the workflow with initial state
initial_state = BMIState(weight=80, height=1.75)

final_state = workflow.invoke(initial_state)
```
